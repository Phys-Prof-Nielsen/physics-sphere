<h2>Problem B1.1</h2>
    </header>

Consider the plots below and assume all measurements are significant to two significant figures.
1. Use the position vs. time graph to estimate that the jet car's velocity t = 20 s. Does it agree with the velocity vs. time graph?
2. Use the velocity vs. time graph that the jet car's acceleration is constant and equal to +5.0 m/s$^2$. 
3. Use the velocity vs. time graph to determine the displacement and compare the result with the displacement from the position vs. time graph.

This problem is a slightly modified version from OpenStax. Access for free at [https://openstax.org/books/college-physics-2e/pages/2-problems-exercises](https://openstax.org/books/college-physics-2e/pages/2-problems-exercises)

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

t = np.arange(0,35,5)

x = 200.0 + 15.0*t + (0.5)*(5.0)*t**2
v = 15.0 + 5.0*t

fig, (ax1,ax2) = plt.subplots(2,figsize=(10,10))

ax1.set(ylabel='x-position (m)', xlabel='Time (s)')

ax1.plot(t,x,'-o')
ax1.set_xlim([0, np.max(t)+0.5])
ax1.set_ylim([0, 4000])
ax1.grid()

ax2.set(ylabel='Velocity (m/s)',xlabel='Time (s)')

ax2.plot(t,v,'-o',color='r')
ax2.set_xlim([0, np.max(t)+0.5])
ax2.set_ylim([0, 180])
ax2.grid()
plt.show()
```

<h2>Problem B1.5</h2>
    </header>

A particle is subject to a net force shown on the graph below.

1. Find the momentum transfer to/from the particle due to this net force.
2. If the particle has a mass of 0.40 kg and travels at +250.0 m/s at t = 5 s, what was its velocity at t = 0 s?
3. What is its velocity at t = 8 s?

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

t1 = np.arange(0,5.1,0.1)
f1 = 0.0 + 3.0*t1
t2 = np.arange(0,5.1,0.1)+5.0
f2 = np.zeros(len(t2))+15

plt.figure()
plt.ylabel('$f_x$ (N)')
plt.xlabel('Time (s)')
ax = plt.gca()
ax.plot(t1,f1,'-',color='b')
ax.plot(t2,f2,'-',color='b')
ax.set_xlim([0, 10])
ax.set_ylim([0, 20])
ax.grid()
plt.show()
```

<h2>Problem B1.6</h2>
    </header>

A particle is subject to a net force shown on the graph below.

1. Find the kinetic energy transfer to/from the particle due to this net force.
2. If the particle has a mass of 0.50 kg and is at rest at t = 0 s, what is its speed at t = 10 s?
3. What is its speed at t = 3 s?

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

t1 = np.arange(0,4.1,0.1)
f1 = -5.0 + 5.0*t1
t2 = np.arange(0,3.1,0.1)+4.0
f2 = np.zeros(len(t2))+15
t3 = np.arange(0,3.1,0.1)+7.0
f3 = 15 - (5/3)*(t3-7)

plt.figure()
plt.ylabel('$f_x$ (N)')
plt.xlabel('Time (s)')
ax = plt.gca()
ax.plot(t1,f1,'-',color='b')
ax.plot(t2,f2,'-',color='b')
ax.plot(t3,f3,'-',color='b')
ax.set_xlim([0, 10])
ax.set_ylim([-5, 20])
ax.grid()
plt.show()
```
