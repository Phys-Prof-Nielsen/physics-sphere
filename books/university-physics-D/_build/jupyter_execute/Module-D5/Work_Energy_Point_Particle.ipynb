## D5.3.1 Review Concept
Together with the <b>Impulse-Momentum Theorem</b>, the <b>Work-Energy Theorem</b> are the cornerstones of physics. While the former is identical to Newton's second law, the later can be derived from Newton's second law. From a conceptual point of view, it is preferable to distinguish the two from Newton's law.

Before we move on, let us review the Impulse-Momentum Theorem (it never hurts to repeat important stuff):

<div class="alert alert-success">
<b>Impulse-Momentum Theorem</b>
    
$$ f_{{net}_x}\Delta t = \Delta p_x \tag{1}$$
</div>

where the impulse is 

<div class="alert alert-success">
<b>Impulse</b>
    
$$ f_{{net}_x}\Delta t = j_x \tag{2}$$
</div>

With that, we can simply write the Impulse-Momentum Theorem as
<div class="alert alert-success">
<b>Impulse-Momentum Theorem again</b>
    
$$ j_x = \Delta p_x \tag{3}$$
</div>

In words: 
<div class="alert alert-success">
A force acting on an object throughout a time interval will change the momentum of the object. <b>The force is causing impulse, which is momentum transfer from one system to another (or within a system). </b>
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D5.3.2 Work-Energy Theorem
Now let us define the Work-Energy Theorem and pay close attention to the similarities to the Impulse-Momentum Theorem.

We define the Work-Energy Theorem as

<div class="alert alert-success">
<b>Work-Energy Theorem</b>
    
$$ f_{{net}_x}\Delta x = \Delta k \tag{4}$$
</div>

where the work is 

<div class="alert alert-success">
<b>Work</b>
    
$$ f_{{net}_x}\Delta x = w \tag{5}$$
</div>

With that, we can simply write the Work-Energy Theorem as
<div class="alert alert-success">
<b>Work-Energy Theorem</b>
    
$$ w = \Delta k \tag{3}$$
</div>

In words: 
<div class="alert alert-success">
A force acting on an object throughout a directed distance will change the kinetic energy of the object. <b>The force is doing work, which is energy transfer from one system to another (or within a system). </b>
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">
<header>
<h3> Example 1</h3>
    </header>

<b>Problem</b>
    <br>
A 2.0-kg box initially at rest is being pushed across the floor by a net force of +25 N. After the box has been pushed a distance of 12 m, how fast is it moving?
   
<b>Methodology</b>
    <br>
Knowing the net force, the directed distance, and that they are in same direction, we can calculate the work. From the work we know the change in kinetic energy. Since the initial kinetic energy is zero, we can then find the final kinetic energy and the associated speed.  

<b>Solution</b>
    <br>
If we let the net force act in the $+x$-direction, then the work is

$$ w = f_{{net}_x}\Delta x = (+25)(+12) = +300~\textrm{Nm}$$

Since the work is energy transfer, the kinetic energy of the box increased by 300 J. We then have:

$$k_f = k_i + \Delta k = (0) + 300 = 300~\textrm{J}$$

Since $k_f = \frac{1}{2}mv^{2}_{f}$, we can solve for the final speed to obtain:

$$v_f = \sqrt{\frac{2k_f}{m}} = \sqrt{\frac{(2)(300)}{(2)}} = 17.32~\textrm{m/s} $$
    
<b>Answer</b>
    <br>
Adjusting for significant figures in our calculations, the final answer is
    
$$ v_f = 17~\textrm{m/s}$$

</div>
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">
<header>
<h3> Example 2</h3>
    </header>

<b>Problem</b>
    <br>
A 1.5-kg box is being pushed across a horizontal floor. At some location, the person pushing the box stops pushing. At that instant the box is moving with a speed of 4.0 m/s. After traveling 3.0 m further, the box comes to a stop. What is the frictional force that causes the box to stop? 
   
<b>Methodology</b>
    <br>
During the last 3.0 m, only friction is acting, so the net work is done by friction only. We know its initial and final speeds, allowing us to find the change in kinetic energy. From this we apply the work-energy theorem to find the work done by the net force (friction in this example). From the definition of work, we can then find the force. 

<b>Solution</b>
    <br>
The change in kinetic energy is

$$ \Delta k = \frac{1}{2}mv_f^2 - \frac{1}{2}mv_i^2 $$
    
Using the given fact that $v_f = 0$, and set the change in kinetic energy equal to the work, we obtain
    
$$ w = - \frac{1}{2}mv_i^2 $$

Now we use the definition for work (Equation 5 above)
    
$$ f_{net,x}\Delta x = - \frac{1}{2}mv_i^2 $$
    
We argued that the net force is the force of friction $f_f$ and we can write the final equation for frictional force as
    
$$ f_{f} = \frac{- \frac{1}{2}mv_i^2}{\Delta x} $$
    
Plugging in our numbers, we get
    
$$ f_{f} = \frac{- \frac{1}{2}(1.5)(4.0)^2}{(3.0)} = -4.0~\textrm{N} $$
    
<b>Answer</b>
    <br>
We notice that the answer is negative, which is a direct consequence of using $\Delta x = +3.0$ m, implying that the frictional force is in the opposite direction of motion.

</div>
