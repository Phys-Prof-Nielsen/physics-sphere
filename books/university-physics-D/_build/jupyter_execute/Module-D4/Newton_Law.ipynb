## D4.2.1 Newton's First Law
It is a conceptual law based on observations:<br>
<div class="alert alert-success">
<b>Newtons 1st Law (N1)</b><br>
    An object in a state of motion will remain in the same state of motion unless an unbalanced external force is acting on the object.
</div>
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.2.2 Newton's Second Law:  Point Particle
The second law is a tool to quantify the first law.

We can proceed to write the relationship between the net force (cause) and the change in momentum (effect) in a form that is most likely more familiar to you, but also simplified due to certain assumptions. We will start with the <b>Impulse-Momentum Theorem</b> where we introduced the concept of net average force (I will omit the "av" subscript now with the understanding that we are dealing with an average force). 

The average force over the interaction time interval will transfer momentum to the particle, and thus change its momentum by:

$$ \Delta p_x = f_{{net}_x}\Delta t $$

and solve for the force, we get:

$$ f_{{net}_x} = \frac{\Delta p_x}{\Delta t} $$

We will now manipulate this equation according to the rules of algebra. Writing the momentum in terms of mass and velocity:

$$ f_{{net}_x} = \frac{\Delta (mv_x)}{\Delta t} $$

and write out the difference in the numerator:

$$ f_{{net}_x} = \frac{mv_{x_2} - mv_{x_1}}{\Delta t}$$

Here, we do not need to have a subscript on the mass since a change in mass makes no sense for a point particle. 
If we follow the logic that a point particle have constant mass (or we can make the assumption that mass is constant), then this reduces to

$$ f_{{net}_x} = m\frac{\left(v_{x_2} - v_{x_1}\right)}{\Delta t} = m\frac{\Delta v_x}{\Delta t}$$

The fraction $\frac{\Delta v_x}{\Delta t}$ is representing <b>the rate of change of velocity</b> and is termed the <b>acceleration</b> of the object. With that in mind and let $a_x$ represent the acceleration, we get

<div class="alert alert-success">
<b>Newtons 2nd Law (N2)</b>
    
$$ f_{{net}_x} = ma_x \tag{1}$$
</div>

<b>which is the most famous version of Newton's 2nd law, but it only applies to a point particle with constant mass</b>.

Another common way of writing Newton's second law is

$$ \sum f_x = ma_x $$

where the $\sum$ symbol means "summing over".

<div class="alert alert-danger">
The acceleration is calculated over a time interval based on to velocities. However, we do not know the <b>instantaneous</b> acceleration at a given time during the time interval. Instead, we can only say something about the <b>average</b> acceleration across that time interval. Similarly, the force is the <b>average</b> net force acting on the particle during this time interval. To gain knowledge of the instantaneous acceleration and force, we must use calculus.
</div>

<header>
<h3> Example 1</h3>
    </header>
    
<b>Problem</b>
    <br>
A particle of mass 0.5 kg is subject to a net average force according to $f_{x,net} = +80.0$ N. The particle is initially moving with a velocity of +2.0 m/s in the positive x-direction.
    
1. Find the average acceleration of the particle during the time period from t = 0 s to t = 5 s.
2. What is the final velocity of the particle?
3. What is the final speed of the particle?
4. What is the instantaenous acceleration at $t = 3.0$ s?
    
<b>Methodology</b>
    <br>
Using the force-acceleration version of Newton's 2nd law, we apply a plug and chuck approch to find the acceleration. From the acceleration, we can obtain the final velocity as we are given the elapsed time and the initial velocity. The speed is simply the magnitude of the velocity since we are restricted to 1D motion. The last question is a conceptual question as we cannot tell anything about the instantaneous acceleration with the information given.

<b>Solution</b>
    <br>
Part (1)<br>
Using Newton's 2nd law (N2), we obtain for the acceleration (dropping the <i>net</i> subscript):
           
$$ a_x = \frac{f_x}{m} = \frac{+80.0}{0.5} = 160~\textrm{m/s}^2 \rightarrow 200~\textrm{m/s}^2 $$

    
Part (2)<br>
Using the definition of acceleration, we have
        
$$a_x = \frac{\Delta v_x}{\Delta t} = \frac{v_{2x} - v_{1x}}{\Delta t} $$
        
Solving for the final velocity (state 2 or event 2):
        
$$ v_{2x} = v_{1x} + a_x \Delta t $$ 
        
Now we insert our numbers and obtain:

$$ v_{2x} = (+2.0) + (160)(5) = +802~\textrm{m/s} \rightarrow +800~\textrm{m/s} $$
    
Part (3)<br>
Since we are moving in a straight line in 1D, the speed is simply the magnitude of the velocity:
        
$$ v = 800~\textrm{m/s} $$
    
Part (4)<br>
We would have to use calculus and have more information than given to say something about the instantaneous acceleration.
        
<b>Evaluation</b>:
In this problem, we have to be careful with the significant figures. Although we can't say anything about the instantaneous acceleration at $t = 3$ s, we know that the average acceleration between $t = 0$ s and $t = 5$ is $a_x = 2~\textrm{E}2~\textrm{m/s}^2$. Given that information, our best guess (assuming a constant profile) would be $a_x = 2~\textrm{E}2~\textrm{m/s}^2$.
