In Newtonian mechanics force is a concept used to describe and quantify interactions between objects. Newton found through observations and experiments that <b>an object's motion would change linearly according to the sum of all external interactions</b> on said object. If there was no net external interaction on the object, its motion would remain unchanged. This came to be known as <b>Newton's first law</b> and it can be divided into two parts: one for linear motion and one for rotational motion.

In Phase D, we will only concern ourselves with a <b>constant net external linear interaction</b> on a single point particle (we should know by now what we mean by a point particle). The discussion on identifying individual interactions is an integral part of Newton's laws and will be explored in Phase C. Furthermore, we are restricted to 1D dynamics in Phase D.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.1.1. Recall: Linear Motion
It is appropriate to re-iterate what we mean by linear motion. According to Newton, the <b>magnitude</b> of linear motion (or momentum) of a point particle is

$$ p = mv $$

where $m$ and $v$ are the mass and speed of the particle, respectively.

The direction associated momentum is specified through the mass and the velocity (directional speed):

$$p_x = mv_x$$

The change in linear momentum in a given direction (say $x$) is the <b>linear impulse</b> between two events 1 and 2:

$$ j_x = \Delta p_x $$

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.1.2 Linear Force on Point Particle 
The net interaction acting on a particle can be described through a <b>net</b> linear force $F_{x_{net}}$ <b>assumed to be constant</b> in this module. This is sometimes referred to as an **unbalanced linear force** if the net linear force is non-zero. The net average interaction acting on the particle over a time interval $\Delta t$ **causes a transfer of momentum** $\Delta p_x$:

<div class="alert alert-success">
    <b>Momentum transfer</b>
    
$$ \Delta p_{x} = f_{{net}_{{av}_{x}}}\Delta t \tag{1}$$
</div>

We can understand this equation as being the mathematical representation of <b>Newton's First Law</b>: 
<div class="alert alert-success">
an object will remain in its same state of motion unless acted upon by a net external force.
</div>

Another verbal representation is: a net interaction acting on an object over an elapsed time is causing the object to experience an impulse or a change in momentum, or the interaction is <b>transferring</b> momentum from one object/system to another. Equation 1 is known as the <b>Impulse-momentum theorem</b>.

### Unit
The standard metric unit of force is kgm/s$^2$, which is commonly defined as a Newton (N).

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">
<header>
<h3> Example 1</h3>
    </header>

<b>Problem</b>
    <br>
A particle, with mass 1.5 kg and intially at rest at t = 0 s, is subject to a net force expressed by $f_x = 2.5$ N. What is the speed of the particle at t = 3 s? Assume exact times.

<b>Methodology</b>
    <br>
Since we know the net force and the elapsed time it is applied to the particle, we can find the impulse (or change in momentum) of the particle. Knowning the initial momentum is zero (at rest initially), we can solve for the final momentum and the velocity speed since mass is given as well.

<b>Solution</b>
    <br>
We can find the impulse from Equation (1):

$$ \Delta p_x = f_{x_{net}}\Delta t $$

Inserting the values provided to us, we get:

$$ \Delta p_x = (2.5)(3.0 - 0) = 7.5~\textrm{kgm/s}$$

While this is the impulse (or change in momentum) of the object, it is not what we were asked to find. The final speed can be found from the impulse since we know the initial speed is zero:

$$ \Delta p_x = p_{x_f} - p_{x_i} = mv_{x_f} - mv_{x_i} $$

Inserting our values:

$$ (7.5) = (1.5)v_{x_f} - (1.5)(0) $$

The final <b>velocity</b> is then

$$ v_{x_f} = +5.0~\textrm{m/s} $$

Since we are asked to solve for the speed, the more appropriate answer is (remember the difference between velocity and speed?)

$$v_f = 5.0~\textrm{m/s}$$

<b>Evaluation</b>
    <br>
The answer tells us that the applied force acting through some time is causing a change in motion of the particle. The net effect is a change in velocity from 0 m/s to +5 m/s.
