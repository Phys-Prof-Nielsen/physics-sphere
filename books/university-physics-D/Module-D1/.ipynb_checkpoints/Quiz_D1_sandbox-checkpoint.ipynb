## Question 3

```python
import random
import numpy as np
import math
from jupyterquiz import display_quiz

cFactor = 1./3.28084

dum = 100.*random.random()

#create the number to convert
digits = random.randint(2,4)
num0 = np.format_float_positional(dum, precision=digits, unique=False, fractional=False, trim='k')
#length of string is digits0+1 due to the decimal point. Check to see if last character is the decimal point
if num0[digits] == '.':
    num0 = math.trunc(float(num0))

x0 = float(num0)*cFactor

#create two correct answers but with different sig figs
answers = []
for ianswers in range(2):
    tmp = np.format_float_positional(x0, precision=digits+ianswers, unique=False, fractional=False, trim='k')
    if tmp[digits+ianswers] == '.':
        answers.append(math.trunc(float(tmp)))
    else:
        answers.append(tmp)
        
#create two incorrect answers but with correct sig figs
scale = [1.05, 1.07]
for ianswers in range(2):
    x0 = scale[ianswers]*float(num0)*cFactor
    tmp = np.format_float_positional(x0, precision=digits, unique=False, fractional=False, trim='k')
    if tmp[digits] == '.':
        answers.append(math.trunc(float(tmp)))
    else:
        answers.append(tmp)
        
#create two incorrect answers (flipped conversion) one with correct sig figs and one with incorrect
x0 = float(num0)/cFactor
for ianswers in range(2):  
    tmp = np.format_float_positional(x0, precision=digits+ianswers, unique=False, fractional=False, trim='k')
    if tmp[digits+ianswers] == '.':
        answers.append(math.trunc(float(tmp)))
    else:
        answers.append(tmp)

example = [{
        "question": "Convert " +str(num0)+ " ft to m",
        "type": "multiple_choice",
        "answers": [
            {
                "answer": answers[3],
                "correct": False,
                "feedback": "False."
            },
            {
                "answer": answers[1],
                "correct": False,
                "feedback": "False."
            },
            {
                "answer": answers[0],
                "correct": True,
                "feedback": "Correct."
            },
            {
                "answer": answers[2],
                "correct": False,
                "feedback": "False."
            },
            {
                "answer": answers[4],
                "correct": False,
                "feedback": "False."
            },
            {
                "answer": answers[5],
                "correct": False,
                "feedback": "False."
            }
        ]
    }]

display_quiz(example)
```
