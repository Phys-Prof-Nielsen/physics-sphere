## D4.1.1 Motivation

We have previously defined motion through **momentum** 

$$\vec{p} = m\vec{v} $$

and then determined **momentum transfer** or **impulse** as a given **force** acting on the particle through time

$$ \vec{i} = \int_{t_1}^{t_{2}}\vec{f}dt $$

The **net** impulse was the **change in momentum** of the particle, which defined the **impulse-momentum theorem**

$$ \int_{t_1}^{t_{2}}\vec{f}_{net}dt = \Delta\vec{p} $$

The impulse-momentum theorem reduces to the more commonly known Newton's 2nd law, where the interactions between two objects are in focus.

<div class="alert alert-info">

We notice that in the above, we must have knowledge of the force as a function of time (unless it is a constant force). At times, we do not have such information. Instead, we have knowledge of the force as a function of postion. In these cases, the work-energy concept is more suitable.
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.1.2 Kinetic Energy

Just as we defined momentum as a measure of motion, kinetic energy is also a measure of motion. We define the kinetic energy of a particle as

<div class="alert alert-success">
    
$$ k = \frac{1}{2}m\vec{v}\cdot\vec{v} $$
</div>

For a system of particles, we can define the center of mass kinetic energy as

<div class="alert alert-success">
    
$$ K = \frac{1}{2}M\vec{V}\cdot\vec{V} $$
</div>

<div class="alert alert-danger">

The kinetic energy of the center of mass is not necessarily the same as the kinetic energy of the system!
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.1.3 Work

Work is a measure of energy transfer to/from a particle by a given force and is defined as the **path integral**

$$ w = \int_L\vec{f}\cdot d\vec{l} $$


<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## D4.1.4 Work-Energy Theorem

The net work is the **net kinetic energy transfer** and is establisged through the **work-energy theorem**

$$ \int_L\vec{f}_{net}\cdot d\vec{l} = \Delta k$$

<div class="alert alert-info">

It should be pointed that the impulse-momentum and work-energy theorems are consistent in the sense that one can be derived from the other. We will do that later in the course. It also leads to a somewhat philosophical question: **what concept is more fundamental? Force or Energy?**. In Newtonian mechanics force is certainly the more fundamental and based on observations. We will later see how a more rigorous energy approach leads to modern field theories and are based on a fundamental principle of nature.
</div>

<div class="alert alert-warning">
<b>Box 2</b><br>
    
A particle is moved around counterclockwise a full rectangular path with vertices at $(x,y) = (0,0), (3,0), (3,2), (0,2)$ m. The force is given by
    
$$ \vec{f} = 3.5xy^2\hat{i} - 5.0y\hat{j} $$
    
What is the work done by this force in moving the particle around the given path. 
</div>

```python
%reset -f

import sympy as sym

x0 = 0.0
y0 = 0.0
x1 = 3.0
y1 = 0.0
x2 = 3.0
y2 = 2.0
x3 = 0.0
y3 = 2.0

x = sym.Symbol('x')
y = sym.Symbol('y')

fx = 3.5*x*y**2
fy = -5.0*y

# Path 1
w1 = sym.integrate(fx,(x,x0,x1)) + sym.integrate(fy,(y,y0,y1))
w1 = w1.evalf(subs={y:y0}) 

# Path 2
w2 = sym.integrate(fx,(x,x1,x2)) + sym.integrate(fy,(y,y1,y2))
w2 = w2.evalf(subs={x:x1}) 

# Path 3
w3 = sym.integrate(fx,(x,x2,x3)) + sym.integrate(fy,(y,y2,y3))
w3 = w3.evalf(subs={y:y2})

# Path 4
w4 = sym.integrate(fx,(x,x3,x0)) + sym.integrate(fy,(y,y3,y0))
w4 = w4.evalf(subs={x:x3})

w = w1 + w2 + w3 + w4

print('Total work done is '+str(w)+' J.')
```
