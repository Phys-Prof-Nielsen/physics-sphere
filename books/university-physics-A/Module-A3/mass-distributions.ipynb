## A3.2.1 Motivation

To know the gravitaional interaction between objects, we will need to know the gravitational field $\vec{g}$ at the location of an object subject to the field. In many cases, the source objects are spherical systems but there may be times where they are not. In those cases, we will need a tool to determine the gravitational field based on the given mass distribution. There are two approaches:

1. Apply the gravitional field from a point mass as obtained from Newton's law of gravity and integrate over the distribution. This is a general approach.

2. Apply Gauss's for gravity. This approach is only quantitatively usefull under certain distributions.

In this note, we will focus on approach 1.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## A3.2.2 Gravitional Field - Point Mass

We saw in previous section that the gravitaional force between two objects were 
    
$$ \vec{F}_G = -\frac{GMm}{r^2}\hat{r} $$

where $\hat{r}$ is a unit vector of $\vec{r}$, which is the **directed line segment from the source mass $M$ to test mass $m$**. 

With the knowledge that the gravitaional force is also given by 

$$ \vec{F}_G = m\vec{g} $$

we can define the **gravitional field** due to a **point mass** $m$ as

<div class="alert alert-success">
    
$$ \vec{g} = -\frac{Gm}{r^2}\hat{r} \tag{1}$$
</div>

Alternatively, and often more useful, we can use the definition of the unit vector:

<div class="alert alert-success">
    
$$ \vec{g} = -\frac{Gm}{r^3}\vec{r} \tag{2}$$
</div>

<div class="alert alert-danger">
    
We must be careful applying the above equations as there is no direct connection to the positions of the source mass and the point of interest.
</div>

A more general approach that directly ties the position vectors into the gravitional field calculation is often preferred. If we use Equation (2) and insert the actual position vectors for the source $\vec{r}'$ and the point of interest $P$ $\vec{r}$:

<div class="alert alert-success">
    
$$ \vec{g} = Gm\frac{\vec{r} - \vec{r}'}{|\vec{r} - \vec{r}'|^3} \tag{3}$$
</div>
    
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## C6.2.3 Solving Strategy

### Set-Up

There are various approaches to finding the gravitational field from a charge distribution. I am presenting one here that I prefer using.

First, let is write down the infinitesimal mass (behave like a point mass) in terms of the mass density and its infinitesimal spatial size:

**Line Charge**

$$ dm = \lambda dl $$

**Surface Charge**

$$ dm = \sigma dA $$

**Volume Charge**

$$ dm = \rho dV $$

The form of the gravitational field equation I will be using is:
   
$$ \vec{g} = -Gm\frac{\vec{r} - \vec{r}'}{|\vec{r} - \vec{r}'|^3} $$


and the field contribution for the infinitesimal mass is 
<div class="alert alert-success">
    
$$ d\vec{g} = -Gdm\frac{\vec{r} - \vec{r}'}{|\vec{r} - \vec{r}'|^3} \tag{4}$$
</div>

### Methodology

Below are the steps taken to obtain the gravivational field. The following sections will illustrate some examples of the approach.

1. Make a sketch. Illustrate the infinitesimal mass, its size, the point P of which we want to know the gravitational field, and positions vectors for the mass as well as the point P.

2. Write down the mass in terms of the density and spatial size (see section above).

3. Write down the position vectors for the infinitesimal mass and the point P. This is usually done in a Cartesian $xyz$-coordinate system but could also be in cylindrical coordinates ($r$, $\theta$, $z$).

4. Write down the distance between the infinitesimal mass and the point P. Express this in terms of the coordinates using Pythagorean theorem.

5. Insert all the information into Equation 4 above.

6. Integrate to obtain the g-field. It is acceptable to look up the integral in a table or use computational methods.

```{admonition} Solution
:class: tip, dropdown

We consider a line mass with charge $Q$ and length $a$ along the x-axis. The task is to find the $g$-field at some point along the x-axis outside the mass distribution.

**Step 1**

Make a sketch.

<img src="A3-fig2.1.png" width="250">


**Step 2**

This is a 1D line mass, so the infinitesimal mass in terms of the density is

$$ dm = \lambda dx' $$

We are given the net mass and the length of the line distribution, so the density is $\lambda = \frac{M}{a}$, and the infinitesimal mass is

$$ dm = \frac{M}{a}dx' $$

**Step 3**

The position vector of the charge is (using 3D Cartesian coordinates):

$$ \vec{r}' = x'\hat{i} + y'\hat{j} + z'\hat{k} $$

All the infinitesimal charges have no $y$ and $z$ components, and we have

$$ \vec{r}' = x'\hat{i}$$

Similarly, the point P has position vector

$$ \vec{r} = x\hat{i}$$

**Step 4**

The distance between the two is

$$ |\vec{r} - \vec{r}'| = (x - x') $$

**Step 5**

The gravitational field from the infinitesimal mass is:

$$ d\vec{g} = -Gdm\frac{\vec{r} - \vec{r}'}{|\vec{r} - \vec{r}'|^3} $$

Insert the information into the equation for the electric field:

$$ d\vec{g} = -\frac{GMdx'}{a}\frac{x\hat{i} - x'\hat{i}}{(x - x')^3} $$

We can factor out the unit vector:

$$ d\vec{g} = -\frac{GM}{a}\frac{(x - x')dx'}{(x - x')^3}\hat{i} $$

Then we divide out the $(x - x')$ factor:

$$ d\vec{g} = -\frac{GM}{a}\frac{dx'}{(x - x')^2}\hat{i} $$

Now we are ready to integrate.

**Step 6**

We integrate to obtain the full gravitational field:

$$ \vec{g} = -\int_{x = 0}^{x = +a}\frac{GM}{a}\frac{dx'}{(x - x')^2}\hat{i} $$

Let us bring all the constants outside the integral, including the unit vector. **The cartesian unit vector is a constant and this step is really important to think about if we are in polar or cylindrical coordinates, where the unit vectors may vary in direction**.

$$ \vec{g} = -\frac{GM\hat{i}}{a}\int_{x' = 0}^{x' = +a}\frac{dx'}{(x - x')^2} $$

The integral can be integrated relatively straightforward by making a substitution $u = (x - x')$. The differential is $du = -dx'$ and we have

$$ \vec{g} = +\frac{GM\hat{i}}{a}\int_{x' = 0}^{x' = +a}\frac{du}{(u)^2} $$

Using the Power Rule for integration:

$$ \vec{g} = -\frac{GM\hat{i}}{a}\frac{1}{u}\bigg\rvert_{x' = 0}^{x' = a} $$

Inserting the expression for u back into the solution:

$$ \vec{g} = -\frac{GM\hat{i}}{a}\frac{1}{(x - x')}\bigg\rvert_{x' = 0}^{x' = a} $$

and evaluate at the limits:

$$ \vec{g} = -\frac{GM\hat{i}}{a}\left(\frac{1}{(x - a)} - \frac{1}{(x)}\right)$$

In an attempt to simplify, we can use common denominator:

$$ \vec{g} = -\frac{GM\hat{i}}{a}\left(\frac{x}{x(x - a)} - \frac{x-a}{x(x-a)}\right)$$

which simplifies to

$$ \vec{g} = -\frac{GM}{x(x - a)}\hat{i}$$

**Step 7**

Several items to reflect on here.

1. There are no $y$ and $z$ components of the gravitional field. This makes sense due to the geometry of the problem.

2. if the point P is far from the line mass, that is, $x >> a$, then we have

$$ \vec{g} = -\frac{GM}{(x)(x)\left(1 - \frac{a}{x}\right)}\hat{i}$$

Since $\frac{a}{x}$ will become very small, it reduces to

$$ \vec{g} = -\frac{GM}{x^2}\hat{i}$$

That is, **far away from the line mass it behaves as a point mass**.
```
