# Linear Force: System of Particles
***

To identify the net force on a system of particles is inherently more complex than just naively extend the tools for a single point particle since each particle in the system may interact with each other. To show the equivalent principle for a system of particle, we must utilize and observation that Newton made and became knowns as Newton's 3rd law:

**For every action there is an equal and opposite re-action**

We can rephrase it as: **for every net interaction on one particle due to a second particle, there is an equal and opposite net interaction on the second particle due to the first particle**.

We will focus on Newton's laws in the next Module.
  
***

## 1. Derivation
Consider a system or an object consisting of $N$ particles labeled by $i$. There are $M$ **external forces** acting on the system with each force labeled as $\vec{F}_k$. In addition, each particle experience an **internal force** due to the other particles. I attempted to illustrate this scenario on the figure below.

<img src="Picture1.png" width="400">

We can consider the net interaction (or net force) on particle $i = 1$ as a sum of internal force and a sum of external forces. The cause and effect (or net force = rate of change of momentum aka Newton's 2nd law) for particle 1 is then

$$ \underbrace{\sum_{j \neq 1}^N \vec{F}_{1j}}_\text{net internal force} + \underbrace{\sum_{k = 1}^M \vec{F}_{1k}}_\text{net external force} = \frac{d\vec{p}_1}{dt}$$

We can write the same for particles $2, 3, \cdots, N$:

$$ \sum_{j \neq 2}^N \vec{F}_{2j} + \sum_{k = 1}^M \vec{F}_{2k} = \frac{d\vec{p}_2}{dt}$$

$$ \vdots $$

$$ \sum_{j \neq N}^N \vec{F}_{Nj} + \sum_{k = 1}^M \vec{F}_{Nk} = \frac{d\vec{p}_N}{dt}$$

where $j$ is the subscript to indicate the interactions due to the $N-1$ particles on the $i$th particle. 

Each equation represents the net interaction on the $i$th particle in the system. The net interaction on the entire system must then be the sum of all these interactions over all the particles:

$$ \sum_{i}^{N}\left(\sum_{j \neq i}^N \vec{F}_{ij} + \sum_{k = 1}^M \vec{F}_{ik} = \frac{d\vec{p}_i}{dt}\right)$$

Distributing the summation through all the terms:

$$ \sum_{i}^{N}\sum_{j \neq i}^N \vec{F}_{ij} + \sum_{i}^{N}\sum_{k = 1}^M \vec{F}_{ik} = \sum_{i}^{N}\frac{d\vec{p}_i}{dt}$$

The second term on the left side of the equation is the **net external force** acting on the entire system: $\vec{F}_{net}^{ext}$. On the right side of the equation, we can move the summation inside the derivative as it does not affect the derivative:

$$ \sum_{i}^{N}\sum_{j \neq i}^N \vec{F}_{ij} + \vec{F}_{net}^{ext} = \frac{d\left(\sum_{i}^{N}\vec{p}_i\right)}{dt}$$

The summation over individual momenta is the total momentum of the system, $\vec{P}$, or the momentum of the system's Center of Mass. We then have:

$$ \sum_{i}^{N}\sum_{j \neq i}^N \vec{F}_{ij} + \vec{F}_{net}^{ext} = \frac{d\vec{P}}{dt}$$

This is getting pretty [exciting](https://www.youtube.com/watch?v=8iwBM_YB1sE)!

A nifty little trick is to use the to use the identity: $\sum_{i}^{N}\sum_{j \neq i}^{N}A_{ij} = \sum_{i<j}^{N}(A_{ij} + A_{ji})$. If you do not know this identity, it is ok and just hit the "I believe button"

<img src="believe.jpg" width="200">

$$ \sum_{i<j}^{N}(\vec{F}_{ij} + \vec{F}_{ji}) + \vec{F}_{net}^{ext} = \frac{d\vec{P}}{dt}$$

If $\vec{F}_{ij}$ is the for acting on particle $i$ due to particl $j$, then $\vec{F}_{ji}$ must be the force acting on particle $j$ due to particle $i$. From Newton's observations, these two forces constitute an action-reaction pair and are equal magnitude and opposite in directions: $\vec{F}_{ij} = -\vec{F}_{ji}$:

$$ \sum_{i<j}^{N}(\vec{F}_{ij} - \vec{F}_{ij}) + \vec{F}_{net}^{ext} = \frac{d\vec{P}}{dt}$$

This is really amazing! The first term on the right side of the equation is then zero, and we have the following result:

<div class="alert alert-success">
$$ \vec{F}_{net}^{ext} = \frac{d\vec{P}}{dt} \tag{1}$$
</div>

<div class="alert alert-success">
    The change on motion of a system of particles is solely determined by the net <b>external</b> force and the system acts as if <b>the net external force is acting at its Center of Mass</b>.
    </div>
    
We can then proceed in a similar manner as we did for the point particle. Let $\vec{P} = M\vec{V}$ be the momentum of the system (as if it is a single particle with mass $M$ located at the CM and moving with a velocity $\vec{V}$). Equation (1) is then

$$ \vec{F}_{net}^{ext} = \frac{d(M\vec{V})}{dt}$$

Using the product rule for derivatives:

$$ \vec{F}_{net}^{ext} = M\frac{d\vec{V}}{dt} + \vec{V}\frac{dM}{dt} $$

If the mass is constant we get

$$ \vec{F}_{net}^{ext} = M\frac{d\vec{V}}{dt} $$

and write it in terms of the linear acceleration of the Center of Mass:

<div class="alert alert-success">
$$ \vec{F}_{net}^{ext} = M\vec{A} \tag{2}$$
</div>

<div class="alert alert-success">
For systems of constant mass, the net external force is causing an acceleration of the CM of the system.
</div>

It is common to ignore the superscript "$ext$" as it is understood using this equation that it is only external forces that are contributing to the change in motion.

<div class="alert alert-info">
We can now see that the tools we developed for a point particle can also be applied to a system and the interpretation is the a system behaves like a point particle with all its mass centered in the CM.
</div>

### Example 1
Famous textbook problems often involves a block, which is treated as a point particle (e.g. see your textbook chapters 4 and 5). We can now understand **why** we can treat it as a point particle since dynamics of the center of mass of the block behaves as a point particle.

Consider a 5.0 kg block moving in the horizontal +x direction on a frictionless surface with an initial speed of 8.0 m/s. Suddenly, it hits a rough segment of length 100 m and is subject to a net resistive frictional force of 2.0 N. 
<ol>
    <li>
        What is the net acceleration of the block on the rough segment?
        <li>
            How far does it travel on the rough segment?
            <li>
                Does the object come to a stop before the rough segment ends?
            <li>
                How long time does it take for the block to come to a stop?
                </ol>

#### Solution
<ol>
    <li>
The frictional force is the only force acting on the block in the horizontal direction and is therefore the only force making up the net force. The mass of the block is constant, so we can use Equation (2) in this case and write it in component form:
        
$$ F_{net,x} = MA_x $$
        
Solving for the acceleration:
        
$$ A_x = \frac{F_{net,x}}{M} $$
        
and inserting the numbers provided:
        
$$ A_x = \frac{(-2.0)}{(5.0)} = -0.40~\textrm{m/s}^2 $$

The block is traveling in the positive direction but experiencing an acceleration in the negative direction, hence it will be slowing down.
    </li>
    
<br>
    
<li>
        The force equation allowed us to find the acceleration of the object, we can now use equations of kinematics to inform about the whereabouts of the object. In this case, we have constant acceleration and we can use Torricelli's equation:<br>
        
$$V^2 = V_{0}^{2} + 2A_{x}\Delta X $$
        
The maximum length it travels on the rough segment is obtained when it comes to a stop, and its final velocity is zero. Solving for the displacement:
        
$$ \Delta X = \frac{V^2 - V_{0}^2}{2A_x} $$
        
and inserting the numbers:
        
$$ \Delta X = \frac{(0)^2 - (8.0)^2}{(2)(-0.40)} = 80~\textrm{m} $$  
    </li>
   <br>
    
<li>
    Since it can only travel 80 m and the rough segment is 100 m long, it does come to a stop before the end of the segment.
    </li>
    <br>
    
<li>
    Since we know the change in velocity and the acceleration (which is constant), we can find the elapsed time:<br>
    
$$ A_x = \frac{\Delta V_x}{\Delta t} $$
    
Solving for the elapsed time:
    
$$ \Delta t = \frac{\Delta V_x}{A_x} $$  
    
and inserting the numbers:
    
$$ \Delta t = \frac{(0) - (8.0)}{(-0.40)} = 20~\textrm{s}$$ 
    
<div class="alert alert-info">
In the above, I used the component form of the vector equations as it turns them into scalar equations and easier to solve. However, we must still pay attention to the directions as components can be either positiove or negative. I also used upper-case letters to emphasize that we are using the Center of Mass parameters. Often, we will not use upper-case letters and simply imply that we are referring to CM dynamics.
</div>
