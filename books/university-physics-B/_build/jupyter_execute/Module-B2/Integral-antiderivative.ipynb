## B2.6.1 The Antiderivative

The integral defined as the antiderivative is considered an inverse derivative. 

<div class="alert alert-success">
    
Simply put, the integral operator undo the derivative operator.
</div>

We use the symbol $\int$ to denote an integral, and the **indefinite integral** of a function $f(x)$ with respect to $x$ is

$$ \int f(x)dx $$

We will discover the *definite integral* in our notes when we consider the integral as an infinite sum.

We notice here: there is a **differential** ($dx$) as part of the integral. The relevance of that will also become clear in the notes when we consider the integral as an infinite sum. For now, we will just kind of ignore it and simply understand it as we are integration with respect to $x$. Wow, that was as an unscientific approach as can be!

Since the integral is the inverse operator of the derivative (it undoes what the derivative did), we can write this statement as

$$ f(x) = \int \frac{d}{dx}[f(x)]dx $$

In this form, we will say the following: **We have knowledge of the derivative of a function, and we desire to find the function itself through integration**.

Let is try and do that through a few examples.

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">       
<header>
<h3> Example 1</h3>
    </header>
    
Consider the function $f(x) = x^2$. Find the integral of $f(x)$.

**Solution**
    
We should ask the question: **which function $F(x)$ will get us $f(x)$ if we take the derivative**?

This may seem like guess work:, but we know the rule for derivative of a polynomial (the power rule). Based on that, if we take a polynomial to the power of $n$, then we decrease the power by $1$ when we take the derivative, so our first guess may be to *increase** the power by $1$ in the function:

$$ F(x) = x^3 $$

If we take the derivative of this, we get

$$ \frac{d}{dx}[F(x)] = 3x^2 $$

This is close to our desired result of $f(x) = x^2$, but we need to get rid of the factor $3$, so we can re-guess a new function

$$ F(x) = \frac{1}{3}x^3 $$

The derivative of that function is

$$ \frac{d}{dx}[F(x)] = x^2 $$

which is our desired result. We would be inclined to say that our integral of $f(x) = x^2$ is

<div class="alert alert-success">
    
$$\int x^{2}dx = \frac{1}{3}x^3 $$
</div>

However, there is a slight element that we are missing here! What if we took the derivative of a function

$$ F(x) = \frac{1}{3}x^3 + C $$

where $C$ is a constant number?

The derivative of constant is zero, so we would have

$$ \frac{d}{dx}[F(x)] = x^2 $$

for that function as well! That is why we call it the **indefinite integral**. There are an infinite number of functions $F(x)$ satisfying the antiderivative as we can add any aribitrary number and still have the same derivative. 

<div class="alert alert-danger">
    
We must add a constant to our result!
</div>

Our final answer is then

<div class="alert alert-success">
    
$$\int x^{2}dx = \frac{1}{3}x^3 + C $$
</div>
</div>
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">       
<header>
<h3> Example 2</h3>
    </header>
    
Consider the function $f(x) = 3x^4$. Find the integral of $f(x)$.

**Solution**
    
We should ask the question: **which function $F(x)$ will get us $f(x)$ if we take the derivative**?

Again, we think of the power rule and based on previous example, we will guess that we must have a polynomial that behaves as

$$ x^5$$

to get $x^4$ after taking the derivative. 

According to the power rule, the 5 is moved down as a factor in front of $x^5$, so we must divide by $5$ to eliminate it. There exists a factor of $3$, which we must maintain, so we should multiply our function by $3$. If we put all that together, we get

$$ F(x) = \frac{3}{5}x^5 $$

Oops, we forgot the constant:

<div class="alert alert-success">
    
$$\int 3x^{4}dx = \frac{3}{5}x^5 + C $$
</div>

**Check**
    
Let us take the derivative to check our answer:

$$ \frac{d}{dx}[F(x)] = \frac{d}{dx}\left[\frac{3}{5}x^5 + C\right] $$

$$ \frac{d}{dx}[F(x)] = \left(\frac{3}{5}\right)\left(5x^4\right) + 0 $$

$$ \frac{d}{dx}[F(x)] =  3x^4 $$

Hooray, we did it!

</div>

## B2.6.2 General Rules 

Based on the above, we can probably guess a general rule regarding polynomials:

1. Consider a function of the form: $f(x) = Ax^r$, where $A$ is a constant and $r$ is a real number. Then we have

<div class="alert alert-success">  
    
$$\int f(x)dx = \frac{A}{r+1}x^{r+1} + C $$
</div>

2. The integral of a sum (or difference) of functions is the same as the sum (or difference) of the integrals

<div class="alert alert-success">
    
$$\int [f(x) + g(x)]dx = \int f(x)dx + \int g(x)dx $$
</div>

3. The integral of a function multiplied by a constant is the same as the constant multiplied by the integral of the function (we have seen that through our examples above)

<div class="alert alert-success"> 
    
$$\int [cf(x)]dx = c\int f(x)dx  $$
</div>

4. The integral of the product of two functions is NOT similar to the product rule of derivatives and is a lot more scary. I am not expecting you to be able to do this one, but may use the result of it. Here it is (where the prime denotes the derivative). This is is known as *integration by parts*.

<div class="alert alert-success"> 
    
$$\int f(x)g'(x)dx = f(x)g(x) - \int g(x)f'(x)dx  $$
</div> 

5. The integral of a composite function is NOT similar to the chain rule of derivatives. In fact, it is a lot more complicated and sometimes does not even exist!

## B2.6.3 Useful Integrals
We have seen how we can find the integral of a polynomial or power law function. We often run into Transcendental functions in physics. The integrals that we will need are (again, we can think of it as undoing the derivative):

<div class="alert alert-success">
    
$$\int \sin(x)dx = -\cos(x) + C  $$
</div>

<div class="alert alert-success">
    
$$\int \cos(x)dx = \sin(x) + C  $$
</div>

<div class="alert alert-success">
    
$$\int \left(\frac{1}{x}\right)dx = \ln(x) + C  $$
</div>

<div class="alert alert-success">
    
$$\int \textrm{e}^x dx = \textrm{e}^x + C  $$
</div>
