Kinematics is an odd child of physics. In it is core, physics attempts to explain **why** the physical world around us behave as it does. Kinematics does not explain any why questions. Instead, it explains **how** concepts of kinematics relates to each other or **how** objects move.

We have alreade seen some of these in our Phase D and C definitions of position, displacement, velocity, and acceleration. Now it is time to generalize them through calculus.

Many textbooks for introductory physics goes into great detail of first considering average velocity and acceleration, and derive equations of kinematics based on those average relationships. My apporach here is more of a analytical approach starting from the definitions of velocity and acceleration.

In this specific notebook, we will obtain expressions for both **general** and **specific** position and velocity functions through indefinite integration.

## B3.1.1 Definitions

Let us recall the mathematical definitions of the kinematic concepts:

* Position vector:

$$\vec{r}$$
* Displacement vector:

$$\Delta\vec{r}$$
* Instantaenous velocity:

$$\vec{v} = \frac{d}{dt}[\vec{r}]$$
* Instantaneous acceleration:

$$\vec{a} = \frac{d}{dt}[\vec{v}]$$

With these definitions, we can derive various relationships, which we call the **equations of motion**.
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

## B3.1.2 General Position Function

We will assume a standard xyz-coordinate system and consider one component of the velocity definition:

$$ v_{x}(t) = \frac{d}{dt}[x(t)] $$

Since the antiderivative undo the derivative, we can perform the following mathematical operation:

$$ \int v_{x}(t)dt = \int\frac{d}{dt}[x(t)] dt$$

$$ \int v_{x}(t)dt = x(t) + \textrm{constant}$$

where I explicitly wrote out the added constant. If we solve for $x(t)$ and define a constant $C_x$ such that

<div class="alert alert-success">
    
$$ x(t) = \int v_{x}(t)dt + C_x \label{eq:1} \tag{1}$$
</div>

then we have an analytical representation of our general position function as a function of time, provided we have knowledge of the velocity function and can calculate the integral.

We can do this for the remaining two position y and z components and summarize it in **vector form**:

<div class="alert alert-success">
    
$$ \vec{r}(t) = \int \vec{v}(t)dt + \vec{C} \label{eq:2} \tag{2}$$
</div>

<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">       
<header>
<h3> Example 1</h3>
    </header>
    
A particle is moving in 1D with a velocity given by $\vec{v}_{x}(t) = [At^2 + Bt]~\hat{i}$, where A = 0.050 m/s$^3$ and B = 1.2 m/s$^2$ . What is the particle's general position function?

**Solution**
    
We identify $v_{x}(t) = 0.050t^2 + 1.2t$ and insert into equation (1):

$$ x(t) = \int v_{x}(t)dt + C_x$$

$$ x(t) = \int (0.050t^2 + 1.2t)dt + C_x $$

$$ x(t) = \frac{0.050}{3}t^3 + \frac{1.2}{2}t^2 + C + C_x $$

where we introduced a constant from the indefinite integration of the velocity. However, a constant added to a constant is just a constant, so we can merge those two constants into one:

<div class="alert alert-success">
    
$$ x(t) = 0.017t^3 + 0.60t^2 + C_x $$
</div>

or in vector form:

<div class="alert alert-success">
    
$$ \vec{r}(t) = [0.017t^3 + 0.60t^2 + C_x]~\hat{i} $$
</div>

<div class="alert alert-info">
    
<b>The Meaning of the Constant C</b><br>
We can find the meaning of the constant $C_x$ by letting t = 0 s. In that case, we have<br>
    
$$ x_0 = C_x $$
    
and $C_x$ then plays the role of the position of the particle at t = 0 s. We call this an <b>initial condition</b>. When $C_x$ is specified, we refer to the solution as a <b>specific solution</b>.
</div>
</div>
<hr style="height:2px;border-width:0;color:gray;background-color:gray">

<div style="background-color: #D0D0D0 ; padding: 10px; border: 1px solid black;">       
<header>
<h3> Example 2</h3>
    </header>
    
A particle is moving in 1D with a velocity given by $\vec{v}_{x}(t) = [At^2 + Bt]~\hat{i}$, where A = 0.050 m/s$^3$ and B = 1.2 m/s$^2$ . At t = 0 s, the particle is located at x = -0.5 m. What is the particle's specific position function?

**Solution**
    
From above, we had

$$ \vec{r}(t) = [0.017t^3 + 0.60t^2 + C_x]~\hat{i} $$

In this example, we are presented with an addtional piece of information: **the initial condition** relating the position of the particle at t = 0 s.

We evaluate the general function at the given time:

$$ x_0 = (0.017)(0)^3 + (0.60)(0)^2 + C_x $$

$$ x_0 = C_x = -0.5~\textrm{m} $$

which tells us that $C_x$ = -0.5 m, and our **specific solution** is then

<div class="alert alert-success">
    
$$ \vec{r}(t) = [0.017t^3 + 0.60t^2 -0.5]~\hat{i} $$
</div>

**NOTE: we can determine the constant with knowledge of a position given at any time. It does not have to be a t = 0 s**.
</div>

## B3.1.3 General Velocity Function

Since the definition of acceleration and velocity are similar, we can extend our analysis above to obtain a general velocity function.

We will assume a standard xyz-coordinate system and consider one component of the acceleration definition:

$$ a_{x}(t) = \frac{d}{dt}[v_{x}(t)] $$
 
Following the same approach as we did for the position case:

$$ \int a_{x}(t)dt = \int\frac{d}{dt}[v_{x}(t)] dt$$

$$ \int a_{x}(t)dt = v_{x}(t) + \textrm{constant}$$

If we solve for $v_{x}(t)$ and define a constant $C_x$ such that

<div class="alert alert-success">
    
$$ v_{x}(t) = \int a_{x}(t)dt + C_x \label{eq:3} \tag{4}$$
</div>

then we have an analytical representation of our general velocity function as a function of time, provided we have knowledge of the acceleration function and can calculate the integral.

We can do this for the remaining two velocity components and write the vector form:

<div class="alert alert-success">
    
$$ \vec{v}(t) = \int \vec{a}(t)dt + \vec{C} \label{eq:4} \tag{4}$$
</div>
