# Newton's Laws
<hr style="border:1px solid gray"> </hr>

Based on observational data Newton established his three law's of motion and the law of gravity. We now know that the true behavior of nature are more complex, and Newton's laws are representations of these complex theories as we restrict ourselves to disucssions involving *everyday* interactions, also known as the **classical limit**.

In this classical regime, Newton's laws are truly remarkable as they allow for a wide variety of not only explanining observed phenomena but also allow for future predictions due to their deterministic form.

We will mostly focus on applying Newton's 2nd law and at times employ the 1st and 3rd laws as conceptual tools.

## Newton's 1st Law: Law of Inertia
<br>
<div class="alert alert-success">
<center><b>An object that is not interacting with anything will remain in its same state of motion</b></center>.
</div>

<img src="rock.jpg" width="400">

This is a conceptual law that can be expressed in many forms. Another way of saying is: if an object is at rest it will remain at rest unless a net external force (or unbalanced force) is acting on it. If an object is moving with some momentum it will remain moving with that momentum unless a net external force (or unbalanced force) is acting on it.

I prefer to use the word "interaction" here as it may help identify individual forces when we try to apply Newton's laws. So, try and get in your mind that an interaction implies that forces are present.

<div class="alert alert-success">
This law also holds true for rotational motion, in which case we replace net force with net torque. 
</div>

<div class="alert alert-danger">
It is possible for the net force to be zero but the net torque to be non-zero. Similarly, it is possible for the net force to ne non-zero but the net torque to be zero. 
</div>

## Newton's 2nd Law
Newton's 2nd law is the tool to quantify the first law.

For linear motion, we have
<br>
<div class="alert alert-success">
$$\vec{F}_{net} = \frac{d\vec{p}}{dt} $$
</div>

If the mass is constant, this reduces to

$$\vec{F}_{net} = m\vec{a} $$


Similarly, for rotational motion

<div class="alert alert-success">
$$\vec{\tau}_{net} = \frac{d\vec{L}}{dt} $$
</div>

and with constant moment of inertia, we have

$$\vec{\tau}_{net} = I\vec{\alpha} $$

<hr style="border:1px solid gray"> </hr>

## Newton's 3rd Law: Action and Re-action
This law states that two interaction objects will experience the same force in magnitude but in opposite directions.

<div class="alert alert-success">
For two objects interacting, object 1 will exert a force $\vec{F}_{12}$ on object 2 and object 2 will exert a force $\vec{F}_{21}$ on object 2 such that<br> 
    <br>
    $$\vec{F}_{12} = -\vec{F}_{21}$$   
</div>

<img src="action.jpg" width="400">

We used this law to show that the Center of Mass motion of a system only depends on the *external forces* acting on the object.

<div class="alert alert-danger">
    The action and re-action forces acts on two <i>different</i> objects and does therefore not appear as a pair of forces when applying Newton's 2nd law on a single object. 
</div>
    
<img src="dwight.jpg" width="400">

This law was actually shown, even before modern physics, to not hold true of all systems. For example, in electricity and magnetism the interaction between charged particles does not obey this law. Instead, if we take the fields themselves into account, then the discrepancy is solved (the reason is that the fields themselves carry momentum).

For our purposes, we will also extend this law to the rotational case and we have that for every torque there is a counter torque.

## Example 1
Consider an object of mass $M = 5.0$ kg subject to forces as shown in the figure. The magnitudes of the applied forces are: $F_1 = 10.0$ N, $F_2 = 3.0$ N, $F_3 = 3.0$ N, and $F_4 = 5.0$ N, where $\vec{F}_1$ makes an angle of $\theta = 15^\circ$ with respect to the horizontal. Find the acceleration of this object.

<img src="fig1.png" width="300">

### Solution
Before we can apply Newton's 2nd law, we must establish a reference frame such that we can assign positive (negative) values to our force components. We have freedom to choose such a frame and in this problem, I have chosen a standard xy-reference system with the origin located as shown.

We now apply Newton's 2nd law (constant mass):

$$ \vec{F}_{net} = M\vec{A} $$

This is a vector equation and we can cast it into component form:

$$ F_{net,x} = MA_x $$

and

$$ F_{net,y} = MA_y $$

which we can solve separately. These equations tells us that only the x-components of the forces contributes to the acceleration in the x-direction and similarly for the the y-components.

**x-component**
We only add the x-components together here. In this case, $\vec{F}_3$ and $\vec{F}_4$ does not contribute, while $\vec{F}_2$ is directed in the negative x-direction and only part of $\vec{F}_1$ contributes. Since we know the angle that $\vec{F}_1$ makes with the positive x-axis, we can use trigonometry to find its x-component. In this case, it is a cosine component. Summing up the forces:

$$ +F_{1}\cos\theta - F_2 = MA_x $$

Solving for the acceleration:

$$ A_x = \frac{+F_{1}\cos\theta - F_2}{M} $$

and inserting values to get:

$$ A_x = \frac{(10.0)\cos(15^\circ) - (3.0)}{(5.0)}  = 1.3~\textrm{m/s}^2$$

**y-component**
In a similar way, we get

$$ -F_{1}\sin\theta  - F_3 + F_4 = MA_y $$

Solving for the acceleration:

$$ A_y = \frac{-F_{1}\sin\theta - F_3 + F_4}{M} $$

and inserting values to get:

$$ A_y = \frac{-(10.0)\sin(15^\circ) - (3.0) + (5.0)}{(5.0)}  = -0.12~\textrm{m/s}^2$$

We conclude that the object is acceleration in the positive x-direction at a rate of $A_x = 1.3~\textrm{m/s}^2$ and in the negative y-direction at a rate of $0.12~\textrm{m/s}^2$.

The acceleration is 

$$\vec{A} = (1.3~\textrm{m/s}^2)~\hat{i} + (-0.12~\textrm{m/s}^2)\hat{j} $$

Since we now know the acceleration of the object, we can say something about its velocity and position provided we have proper initial information such as initial velocity and initial position.

## Example 2
Consider a solid cylinder of mass $M = 2.0$ kg and radius $R = 0.3$ m. Two forces of equal magnitude of $15$ N are applied as shown in the figure. The cylinder has its long axis along the z-direction, so the figure illustrate the xy-plane. The z-axis is positive into the screen. Find angular acceleration of the cylinder about the z-axis.

<img src="fig2.png" width="300">

### Solution
We must find the net torque acting on the object. A positivbe torque (in the +z direction) will correspond to a clockwise rotation so I added a directed arc element to illustrate that clockwise rotation is considered positive in this case. We then see that $\vec{F}_1$ will cause a rotation in the positive direction and likewise for $\vec{F}_2$. Hence, both applied forces will cause positive torques. We could also use the right-hand rule by using the directed line segment from the z-axis to the point of interaction.

In the following, we will use:

**Moment of inertia of a cylinder**
$$ I = \frac{1}{2}MR^2 $$

**Magnitude of torque due to a force**
$$ \tau = rF\sin\theta $$

In our case, $r$ is the magnitude of the directed line segment from the z-axis to the point of interaction and we can identify that as the radius R for both forces. The angle between this directed line segment and the applied force is in both cases 90$^\circ$.

Adding up the individual torques:

$$ +\tau_1 + \tau_2 = I\alpha $$

$$ r_{1}F_{1}\sin\theta_1 + r_{2}F_{2}\sin\theta_2 = I\alpha $$

Replacing the variable names with the ones identified:

$$ RF_{1}\sin\theta_1 + RF_{2}\sin\theta_2 = \frac{1}{2}MR^2\alpha $$

Now solve for the angular acceleration:

$$ \alpha = \frac{RF_{1}\sin\theta_1 + RF_{2}\sin\theta_2}{\frac{1}{2}MR^2} $$

and inserting our values:

$$ \alpha = \frac{(0.3)(15)\sin(90^\circ) + (0.3)(15)\sin(90^\circ)}{\frac{(1)}{(2)}(2.0)(0.3)^2} = +100~\textrm{rad/s}^2$$

The cylinder is rotating in a positive direction (clockwise) and increasing its angular speed at a rate of 100 rad/s$^2$.

<div class="alert alert-info">
NOTE: We see that the forces add up to zero and the Center of Mass of the cylinder does not have any linear acceleration. The cylinder is purely rotating. This is an example of having zero net force but non-zero net torque.
</div>
