## 1. Area

We have seen that we can define the net interaction on a point particle through 

$$ \Delta\vec{p} = \int_{t_1}^{t_2}\vec{F}_{net}dt \tag{1}$$

and if the applied net force is constant (or considering the average force over the time interval):

$$ \Delta\vec{p} = \vec{F}_{net}\Delta t \tag{2}$$

and for a system of particles, where we consider the Center of Mass of the system:

$$ \Delta\vec{P} = \int_{t_1}^{t_2}\vec{F}_{net}^{ext}dt \tag{3}$$

and if the applied net force is constant (or considering the average force over the time interval):

$$ \Delta\vec{P} = \vec{F}_{net}^{ext}\Delta t \tag{4}$$

where *ext* refers to external forces acting on the system.

In any of these cases, we can interpretate it as the following:

<div class="alert alert-success">
    The impulse (or change in momentum) of a point particle or a center of mass system is the <b>area</b> bounded by the graph and the time-axis in a net force vs. time graph.
</div>

***

### Example 1
A 2.0 kg point particle is subject to a net force vs. time graph as the one shown by executing the code below. Assume all interactions takes place along the x-direction. 
<ol>
    <li>What is the linear impulse of the particle over the time interval $t = [0,10]$?
        <li>If the particle is at rest at t = 0 s, what is its velocity at t = 10 s?
</ol>

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

t1 = np.arange(0,5.1,0.1)
F1 = 2.0*np.ones(len(t1))
t2 = np.arange(0,5.1,0.1) + 5.0
F2 = -0.4*t2 + 4.0

plt.figure()
plt.ylabel('Net Force (N)')
plt.xlabel('Time (s)')
ax = plt.gca()
ax.plot(t1,F1,'b')
ax.plot(t2,F2,'b')
ax.set_xlim([0, 10.5])
ax.set_ylim([0, 2.5])
ax.grid()
plt.show()
```

#### Solution
<ol>
    <li>The impulse is the area bounded by the graph and the time axis. We can divide it into two segments: a rectangle and a triangle. The ara of the rectangle is $A_1 = (2.0)(5) = 10$ Ns and the area of the triangle is $A_2 = (0.5)(5)(2.0) = 5.0$ Ns. The total area, and the impulse, is then

$$\vec{J} = (15~\textrm{Ns})~\hat{i}$$
        
NOTE: the unit Ns is the same as kgm/s.
        
<li>The impulse is the change in momentum of the particle:
    $$ m\vec{v}_2 - m\vec{v}_1 = \vec{J} $$
    
 Solving for the final velocity:
    $$\vec{v}_2 = \vec{v}_1 + \frac{\vec{J}}{m} $$   
    and inserting our numbers:   
    $$\vec{v}_2 = (0)~\hat{i} + \frac{(15)~\hat{i}}{(2.0)} = (7.5~\textrm{m/s})~\hat{i} $$

## 2. Slope
Through the relationships

$$ \vec{F}_{next} = \frac{d\vec{p}}{dt} $$

and 

$$ \vec{F}_{next}^{ext} = \frac{d\vec{P}}{dt} $$

we observe that 

<div class="alert alert-success">
    The net force is represented by the slope of a momentum vs. time graph.
</div>

This relationship is rarely practical, and is just something we should keep in the back of our mind in case we stumble upon it.
