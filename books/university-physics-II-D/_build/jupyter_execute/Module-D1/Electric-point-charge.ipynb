## D1.2.3 The Elementary Charge and Unit of Charge
The smallest amount of charge that can exists is known as the **elementary charge** and has a value of $e = 1.602176634\times 10^{-19}~\textrm{C}$, where Coulomb (C) is the SI unit of electrical charge. The smallest electric charge particle we know of without size and shape is the electron, and the electron therefore has a magnitude charge of $e$. Since an atom is electrically neutral (no net charge), the proton must also have a charge magnitude of $e$. By convention, the charge of a proton is $+e$ while the charge of the elecron is $-e$.

## D1.2.4 Experimental Properties of Electrical Charges

1. Two types exist: positive and negative (these terms were named by Benjamin Franklin).
2. Like charges repel and unlike charges attract one another.
3. Charge is a conserved quantity.
4. All non-elementary particles have electrical charge in multiple of $e$.
4. Some elementary particles, like quarks, have electrical charge in fractions of $e$.

Although quarks have fractional charge of $e$, they do not appear naturally (or free) in nature but always come in sets that make up integer multiple charge of $e$.

A large tank capacitor used in a Tesla coil is charged to a net charge of $2.0~\textrm{mC}$. If this excess charge is caused by electrons, approximately how many extra electrons are placed on the capacitor compared to its neutral state?

```python
from jupyterquiz import display_quiz
import json
with open("quiz-electric-charge.json", "r") as file:
    questions=json.load(file)

display_quiz(questions)
```

### Box 1 Solution

```python
from math import floor  #import for rounding down to integer

# charge of a single electron
e = 1.602E-19

# total charge on capacitor
q = 2.0E-3

# number of electrons: N*e = q
N = q/e

# in case it is a decimal number, we can round down using the floor function
print('Number of electrons: '+str(floor(N)))
```
