# Introduction for Formatting Markdown Boxes in Jupyter
The purpose of this little document is to provide you with some of the basic features of formatting Jupyter markdown boxes. A lot of this can be found from [this webpage](https://medium.com/analytics-vidhya/the-ultimate-markdown-guide-for-jupyter-notebook-d5e5abf728fd).

I recommend you spend some time 1-2 hours going through it, and making yourself familiar with the formatting. Try and make your own document, etc.
